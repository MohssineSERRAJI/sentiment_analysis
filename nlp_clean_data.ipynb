```
import pandas as pd
import numpy as np
import nltk
from nltk.stem.porter import PorterStemmer 

#this step for download the stopwords
from nltk.corpus import stopwords
nltk.download('stopwords')
STOPWORDS = set(stopwords.words('english'))
#load data
df = pd.read_table("tweets.txt")
#Remove na values form tweets
df['3text'].fillna("", inplace= True)

#the function used to clean data one line
def clean_data(text):
  '''
    this function used to clean data
    - remove punctuation
    - remove stop words
    - convert text to lower
    - tokenize the text
    - stemming the text
  '''
  #create the stemmer
  stemmer = PorterStemmer()
  tokenizer = nltk.RegexpTokenizer(r"\w+")
  text = tokenizer.tokenize(text)
  #convert to lower than stem the words
  text_without_ws = [stemmer.stem(word.lower()) for word in text if not word in STOPWORDS]
  return text_without_ws

new_df = df["3text"].apply(clean_data)
print(new_df[0])
```
